```python
# WHILE loop:-

# Print values 1 to 10:-

# DECLARE X: INTEGER
# SET i:= 1
# WHILE i<=10
#    PRINT i
#    INCREMENT i
# END WHILE

i = 1
while i<=10:
    print(i)
    i+=1
```

```python
# TABLE:-

   # DECLARE num : INTEGER 
   # INTEGER i
   # SET i=1
   # READ num
   # WHILE i <= 10
   #   PRINT  num * i
   #   INCREMENT i
   # END WHILE

num = int(input("enter:"))
i = 1
while i<=10:
    print(num, "x" , i , "=" ,num * i)
    i += 1
```

```python
# sum of the series: 1+2+3+......+n :-

# DECLARE i: INTEGER
# DECLARE N: INTEGER
# DECLARE sum: INTEGER
# SET i:=1
# SET sum = 0
# WHILE i<=n
#     sum:= sum + i
#     INCREMENT i
# ENdD WHILE
# PRINT sum
# PRINT sum/n

n = int(input("enter the number:"))
i = 1
sum = 0
while i <= n:
    sum = sum + i
    i += 1
print("Sum of the series:" , sum)
print("Average of the series:" , sum/n)

```

```python
# sum of the series: 1 + 2^2 + 3^3 +......+ n^2 :-

# DECLARE i: INTEGER
# DECLARE N: INTEGER
# DECLARE sum: INTEGER
# SET i:=1
# SET sum = 0
# WHILE i<=n
#     sum:= sum + i*i
#     INCREMENT i
# ENdD WHILE
# PRINT sum
# PRINT sum/n

n = int(input("enter the number:"))
i = 1
sum = 0
while i <= n:
    sum = sum + i*i
    i += 1
print("Sum of the series:" , sum)
print("Average of the series:" , sum/n)
```

```python
# FIBONACCI SERIES :-

    # DECLARE n : INTEGER
    # DECLARE prev : INTEGER
    # DECLARE Next : INTEGER
    # DECLARE num : INTEGER
    # SET prev := 0
    # SET next := 1
    # SET sum := 0
    # PRINT prev
    # PRINT next
    # WHILE sum <= n
    #    sum := prev + next
    #    PRINT sum
    #    prev := next
    #    next := sum
    # END WHILE

n = int(input("Enter the number:"))
prev = 0
next = 1
sum = 0
while sum <= n:
    sum = prev + next
    print(sum)
    prev = next
    next = sum
```

```python
# Palindrome Digits 
   # Declare n : Integer 
   # Declare num : Integer 
   # int b : Integer
   # int a : Integer
   # Set b = 0
   # Read num
   # while n>0 
   #    a = n MOD 10
   #    n = n / 10
   #    b = b * 10 + a
   # end while
   # if b == num then
   #   print "Palindrome"
   # else 
   #   print "Not Palindrome"
   # end if


num = int(input("enter: "))
b = 0
n = num

while n > 0:
    a = n % 10
    n = n // 10   
    b = b * 10 + a

if b == num:
    print("Palindrome")
else:
    print("Not Palindrome")
    

```

```python
#  Armstrong no :-

# Declare a : Integer 
 # Declare b : Integer
 # Declare n : Integer
 # Declare num : Integer
 # Set b := 0
 # Read num
 # n := num
 # While n>o
 #   a := n MOD 10
 #   n := n/10
 #   b := b+a*a*a
 # end while
 # if b == num then   
 #    print "Armstrong"
 # else 
 #    print "not armstrong"
 # end if 

num = int(input("enter: "))
b = 0
n = num

while n > 0:
    a = n % 10
    n = n // 10   
    b = b + a*a*a

if b == num:
    print("Armstrong")
else:
    print("Not Armstrong")
    
```

```python
#  FOR loop:-

# print the values 1 to 10:-

  # DECLARE num : INTEGER
  # READ Num
  # FOR i=1 TO 10 step 1
  #    PRINT num * i
  # END FOR

num = int(input("enter:"))
for i in range(1,num+1):
    print(i)
```

```python
# sum of the series: 1+ 1/2 + 1/3 +......+ 1/n :-

# DECLARE N: INTEGER
# DECLARE sum: REAL
# SET sum = 0.0
# READ n
# FOR i:= 1 TO n STEP 1
#     sum:= sum + 1/i
# END FOR
# PRINT sum
# PRINT sum/n

n = int(input("enter the number:"))
sum = 0
for i in range(1,n):
    sum = sum + 1/i
print("Sum of the series:" , sum)
print("Average of the series:" , sum/n)
```

```python
# sum of the series: 1+ 1/2^2 + 1/3^2 +......+ 1/n^2 :-

# DECLARE N: INTEGER
# DECLARE sum: REAL
# SET sum = 0.0
# READ n
# FOR i:= 1 TO n STEP 1
#     sum:= sum + 1/i^2
# END FOR
# PRINT sum
# PRINT sum/n

n = int(input("enter the number:"))
sum = 0
for i in range(1,n):
    sum = sum + 1/i**2
print("Sum of the series:" , sum)
print("Average of the series:" , sum/n)

```
