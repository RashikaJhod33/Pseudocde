# Find Cube:-

```python
# FUNCTION Cube(num : INTEGER) RETURN INTEGER
#           RETURN num * num *num
# END FUNCTION
# PRINT Cube(5)

def cube(num):
    return num*num*num
print("5^3 =",cube(5))
```

# Absalute Value:-

```python
 # FUNCTION abs(x: INTEGER): INTEGER
 #    IF x>0 THEN 
 #       RETURN x
 #    ELSE 
 #       RETURN -x
 #  END FUNCTION
 #  PRINT abs(5)
 #  PRINT abs(-5)

def abs(x):
    if x>0:
        return x
    else:
        return -x
print("Absalute Value of 5 is",abs(5),".")
print("Absalute Value of -5 is",abs(-5),".")
```

# Factorial:-

```python
# FUNCTION factorial(n : INTEGER): INTEGER
#     DECLARE fact : INTEGER
#     SET fact = 1
#     FOR i in 1 TO n
#       fact = fact * i
#     END FOR
#     RETURN fact
# END FUNCTION 
# PRINT factorial(5)

def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact = fact * i
    return fact

num = int(input("Enter a number: "))
print("Factorial =", factorial(num))
```

# Sum the series 1....n:-

```python
# FUNCTION sum(n: Integer): Integer
#     DECLARE s:= Integer
#     SET s:= 0
#     FOR i:=1 TO n
#         s:= s+i
#     END FOR
#     RETURN S
# END FUNCTION
# PRINT sum(4)

def sum(n):
    s = 0
    for i in range(1, n+1):
        s = s + i
    return s
print(sum(4))
        
```

# Sum of Series 1+1/2+1/3+1/4....1/n:-

```python
# FUNCTION sum(n: INTEGER): REAL
#     DECLARE s: REAL
#     SET s := 0
#     FOR i=1 TO n
#          s: s+1/i
#     END FOR
#     RETURN s
# END FUNCTION
# PRINT sum(4)

def sum(n):
    s = 0
    for i in range(1, n+1):
        s = s + 1/i
    return s
print(sum(4))
```

# Reverse Number :-

```python
 # FUNCTION reverse(n : INTEGER): INTEGER
 #    DECLARE a: INTEGER
 #    DECLARE b : INTEGER
 #    SET b := 0
 #    WHILE n>o LOOP
 #      a:= n MOD 10
 #      n:= n/10
 #      b:= b*10+a
 #    END WHILE
 #    RETURN b
 # PRINT reverse(12321)

def reverse(n):
    b = 0
    while n>0:
        a = n%10
        n = n//10
        b = b*10+a
    return b
print(reverse(123))
```

# Palindrome:- 

```python
# FUNCTION Palindrome(n: INTEGER): STRING
#   DECLARE a : INTEGER
#   DECLARE b : INTEGER
#   DECLARE c : INTEGER
#   SET b := 0
#   READ x
#   n := x
#   WHILE n>0
#     a:= n MOD 10
#     n:= n/10
#     b:= b*10+a
#   END WHILE
#   IF b==x THEN
#     RETURN "Palindrome"
#   ELSE 
#     RETURN "not Palindrome"
#   END IF
# END FUNCTION
# PRINT Palindrome(121)

def palindrome(n):
    b = 0
    x = n  
    
    while n > 0:
        a = n % 10
        n = n // 10
        b = b * 10 + a
        
    if b == x:
        return "Palindrome"
    else:
        return "Not Palindrome"

print(palindrome(12321))
```
