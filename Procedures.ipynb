 # Simple Interest:-

```python
# PROCEDURE sim_Int (P: INTEGER, R: INTEGER, Y: INTEGER) 
#         DECLARE SI = REAL
#         SI := (P * R * Y)/100
#         PRINT SI
# END PROCEDURE
# CALL sim_Int(1000,5,2)

def sim_Int(P, R, Y):
    SI = (P * R * Y) / 100
    print("Simple Interest:",SI)

sim_Int(1000, 5, 2)
```

# Find Cube :-

```python
# PROCEDURE cube (num: INTEGER)
#      DECLARE cube : INTEGER
#      cube : num * num * num
#      PRINT cube
# END PROCCEDURE
# CALL cube(5)

def cube(num):
    cube = num * num * num
    print("5^3=",cube)
cube(5)
```

# Find area of triangle :-

```python
# PRCEDURE AreaOfTriangle ( b: REAL, h: REAL)
#      PRINT 0.5 * b * h
# END PROCEDURE
# CALL AreaOfTriangle(4.5, 2.5) 

def AreaOfTriangle(b,h):
    print("Area of Traingle:",0.5*b*h)
AreaOfTriangle(4.5 , 2.5)
```

# Find maximum value:-

```python
# PROCEDURE max(a: INTEGER, b: INTEGER)
#      IF a>b
#        PRINT a
#      ELSE 
#        PRINT b
# END PROCEDURE
# CALL max(10, 5)

def max(a,b):
    if a>b:
        print(a,"is greater than" , b,".")
    else:
        print(b,"is greater than",a)
max(10,5)
```

# Find minimum value:-

```python
# PROCEDURE min(a: INTEGER, b: INTEGER)
#      IF a<b
#        PRINT a
#      ELSE 
#        PRINT b
# END PROCEDURE
# CALL min(10, 5)

def min(a,b):
    if a<b:
        print(a,"is less than" , b,".")
    else:
        print(b,"is less than" , a,".")
min(10,5)
```
