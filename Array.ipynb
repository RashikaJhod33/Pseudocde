```python

# MAXIMUM VALUE

# DECLARE a : ARRAY[0:4] OF INTEGER
# DECLARE max: INTEGER 
# FOR i:=0 TO 4
#     READ a[i]
# END FOR
# SET max := a[0]
# FOR i:=0 RO 4
#     IF a[i] > max THEN 
#         max := a[i]
# END FOR
# PRINT max

a = [] 
for i in range(5):
    num = int(input("Enter number: "))
    a.append(num)
max = 0
for i in range(5):
    if a[i] > max:
        max = a[i]
print("maximum:",max)
```

```python
# MINIMUM VALUE

# DECLARE a : ARRAY[0:4] OF INTEGER
# DECLARE min: INTEGER 
# FOR i:=0 TO 4
#     READ a[i]
# END FOR
# SET max := a[0]
# FOR i:=0 RO 4
#     IF a[i] < min THEN 
#         min := a[i]
# END FOR
# PRINT min

a = [] 
for i in range(5):
    num = int(input("Enter number: "))
    a.append(num)
min = 0
for i in range(5):
    if a[i] < min:
        min = a[i]
print("Minimum :", min)
```

```python
# Multidimentional array:-
# DECLARE X: ARRAY[2][2] OF INTEGER
# FOR i:=0 TO 1
#     FOR j:=0 TO 1
#        READ X[i][j]
#     END FOR
# END FOR
# FOR i:=0 TO 1
#     FOR j:=0 TO 1
#        PRINT X[i][j]
#     END FOR
# END FOR


X = []
for i in range(2):
    row = []
    for j in range(2):
        num = int(input(f"Enter element at position [{i}][{j}]: "))
        row.append(num)
    X.append(row)
for i in range(2):
    for j in range(2):
        print(X[i][j], end=" ")
    print()
```

```python
# Sum of elements in 2d array:-
# DECLARE X: ARRAY[2][2] OF INTEGER
# DECLARE sum: INTEGER
# SET sum:=0
# FOR i:=0 TO 1
#     FOR j:=0 TO 1
#        READ X[i][j]
#     END FOR
# END FOR
# FOR i:=0 TO 1
#     FOR j:=0 TO 1
#       sum:= sum + X[i][j]
#     END FOR
# END FOR
# PRINT sum

X = []
sum= 0
for i in range(2):
    row = []
    for j in range(2):
        num = int(input(f"Enter element at position [{i}][{j}]: "))
        row.append(num)
    X.append(row)
for i in range(2):
    for j in range(2):
        sum = sum + X[i][j]
print("Sum =", sum)
```

```python
# Addition of matrix:-
# DECLARE X: ARRAY[3][3] OF INTEGER
# DECLARE Y: ARRAY[3][3] OF INTEGER
# DECLARE Z: ARRAY[3][3] OF INTEGER
# FOR i:=0 TO 2
#     FOR j:=0 TO 2
#        READ X[i][j]
#        READ Y[i][j]
#     END FOR
# END FOR
# FOR i:=0 TO 2
#     FOR j:=0 TO 2
#        Z[i][j] = X[i][j] + Y[i][j]
#     END FOR
# END FOR

X = []
Y = []
Z = []

print("Enter elements of Matrix X:")
for i in range(3):
    row = []
    for j in range(3):
        num = int(input(f"X[{i}][{j}]: "))
        row.append(num)
    X.append(row)

print("Enter elements of Matrix Y:")
for i in range(3):
    row = []
    for j in range(3):
        num = int(input(f"Y[{i}][{j}]: "))
        row.append(num)
    Y.append(row)

for i in range(3):
    row = []
    for j in range(3):
        row.append(X[i][j] + Y[i][j])
    Z.append(row)

print("Resultant Matrix:")
for i in range(3):
    for j in range(3):
        print(Z[i][j], end=" ")
    print()
```

```python
# Count how many values are even or odd in 3*3 matrix:-
# DECLARE X: ARRAY[3][3] OF INTEGER
# DECLARE even : INTEGER
# DECLARE odd : INTEGER 
# SET even = 0 , odd = 0
# FOR i:=0 TO 2
#     FOR j:=0 TO 2
#        READ X[i][j]
#     END FOR
# END FOR
# FOR i:=0 TO 2
#     FOR j:=0 TO 2
#        IF X[i][j] MOD 2 == 0 THEN
#            INCREMENT even
#        ELSE
#            INCREMENT odd
#        END IF
#     END FOR
# END FOR
# PRINT even
# PRINT odd

X = []

even = 0
odd = 0
for i in range(3):
    row = []
    for j in range(3):
        num = int(input(f"Enter element at [{i}][{j}]: "))
        row.append(num)
    X.append(row)
for i in range(3):
    for j in range(3):
        if X[i][j] % 2 == 0:
            even += 1
        else:
            odd += 1

print("Even numbers =", even)
print("Odd numbers =", odd)
```
